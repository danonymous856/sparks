To predict the percentage of students based on their study hours

```python
import pandas as pd
```

```python
data = pd.DataFrame({'Hours':[2.5,5.1,3.2,8.5,3.5,1.5,9.2,5.5,8.3,2.7,7.7,5.9,4.5,3.3,1.1,8.9,2.5,1.9,6.1,7.4,2.7,4.8,3.8,6.9,7.8],
'Scores':[21,47,27,75,30,20,88,60,81,25,85,62,41,42,17,95,30,24,67,69,30,54,35,76,86]
})
```

### Time to plot 

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('Solarize_Light2')
fig,ax = plt.subplots()
ax.set_title("Study Hours Based Percentage")
ax.scatter(data['Hours'],data['Scores'],color = 'g',marker = 'o')
ax.set_xlabel('Time')
ax.set_ylabel('Percentage')
plt.show()
```

```python
# As the data set is quite small spliting the data won't give us much benificiary improvement of the model.
```

```python
# Importing Scikit-learn and numpy, selecting  a linear model
import sklearn.linear_model
import numpy as np
```

```python
# Prepare the data
X = np.c_[data['Hours']]
y = np.c_[data['Scores']]
```

```python
model = sklearn.linear_model.LinearRegression()
```

```python
# Train the model
model.fit(X,y)
```

```python
X_new = [[9.25]] #This is the study hours of the student
print(model.predict(X_new)) #Here we predict the score of that student
```
